## Prepare Data for Time Series Modeling

To prepare the dataset for time series modeling, I will first convert the 'Year' column to datetime objects, then set it as the DataFrame's index, and finally sort the DataFrame by this new index to ensure chronological order. This aligns with the instructions provided for the subtask.



```python
import pandas as pd

# Convert 'Year' column to datetime objects and set as index
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df = df.set_index('Year')

# Sort the DataFrame by its new datetime index
df = df.sort_index()

print("DataFrame after setting 'Year' as datetime index and sorting:")
display(df.head())
print("\nDataFrame Info:")
df.info()
```

## Split Data into Training and Testing Sets

### Subtask:
Divide the dataset into training and testing sets to evaluate the model's performance on unseen data. A common approach for time series is to use an earlier period for training and a later period for testing.


**Reasoning**:
To prepare for time series modeling, I need to split the data into training and testing sets. I will define a split point that allocates 80% of the data for training and 20% for testing, create corresponding DataFrames, and then separate the target variable ('Growth Rate') and features (the 'Year' index) for both sets.



```python
import numpy as np

# Define the split point (e.g., 80% for training, 20% for testing)
split_index = int(len(df) * 0.8)

# Create training and testing DataFrames
train_df = df.iloc[:split_index]
test_df = df.iloc[split_index:]

# Separate target variable (y) and features (X)
X_train = train_df.index.values.reshape(-1, 1)
y_train = train_df['Growth Rate']

X_test = test_df.index.values.reshape(-1, 1)
y_test = test_df['Growth Rate']

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_test: {y_test.shape}")

print("\nFirst 5 rows of train_df:")
display(train_df.head())

print("\nFirst 5 rows of test_df:")
display(test_df.head())
```

## Train a Time Series Forecasting Model

### Subtask:
Select and train an appropriate time series forecasting model (e.g., a simple linear regression using time as a feature, or more advanced models like ARIMA/Prophet if deemed necessary based on preliminary analysis) to predict the 'Growth Rate'.


**Reasoning**:
To prepare the data for linear regression, I will convert the datetime objects in `X_train` and `X_test` to numerical year integers. Then, I will import and instantiate a `LinearRegression` model from `sklearn.linear_model` and fit it using the numerical training data.



```python
from sklearn.linear_model import LinearRegression

# Convert X_train and X_test from datetime objects to numerical years
X_train_numeric = pd.to_datetime(X_train.flatten()).year.values.reshape(-1, 1)
X_test_numeric = pd.to_datetime(X_test.flatten()).year.values.reshape(-1, 1)

# Instantiate the Linear Regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train_numeric, y_train)

print("Linear Regression model trained successfully.")
```

**Reasoning**:
Now that the linear regression model has been trained, I will use it to make predictions on the test data and then evaluate its performance using common regression metrics such as Mean Absolute Error (MAE), Mean Squared Error (MSE), and R-squared to assess how well the model generalizes to unseen data.



```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Make predictions on the test set
y_pred = model.predict(X_test_numeric)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R2): {r2:.4f}")
```

**Reasoning**:
To visually assess the model's performance and understand the nature of the predictions, especially given the low R-squared value, I will create a plot comparing the actual 'Growth Rate' from the test set with the model's predicted 'Growth Rate' values.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 7))
sns.lineplot(x=test_df.index, y=y_test, label='Actual Growth Rate', color='blue')
sns.lineplot(x=test_df.index, y=y_pred, label='Predicted Growth Rate', color='red', linestyle='--')
plt.title('Actual vs. Predicted Growth Rate Over Time (Test Set)')
plt.xlabel('Year')
plt.ylabel('Growth Rate')
plt.legend()
plt.grid(True)
plt.show()
```
